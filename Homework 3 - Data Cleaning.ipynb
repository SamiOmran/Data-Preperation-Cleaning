<h2> First get missing values from factors column </h2>

```python
factors = {
    'eco': 'Economy (GDP per Capita)',
    'fam': 'Family',
    'health': 'Health (Life Expectancy)',
    'free': 'Freedom',
    'trust': 'Trust (Government Corruption)',
    'gen': 'Generosity',
    'dys': 'Dystopia Residual'
}
```

```python
df[(df['Health (Life Expectancy)'].isna()) & (df['Happiness Score'].notna())].head(3)
```

```python
def calculate_missing(score, factors):
    '''
        This function takes the Happiness score value and the factors as list
        and returns the missing value
    '''
    return score - sum(factors)
```

```python
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].fillna(
    calculate_missing(df['Happiness Score'], [
        df[factors.get('eco')],
        df[factors.get('fam')],
        df[factors.get('free')],
        df[factors.get('trust')],
        df[factors.get('gen')],
        df[factors.get('dys')]
    ]))
```

```python
df['Economy (GDP per Capita)'] = df['Economy (GDP per Capita)'].fillna(
    calculate_missing(df['Happiness Score'], [
        df[factors.get('health')],
        df[factors.get('fam')],
        df[factors.get('free')],
        df[factors.get('trust')],
        df[factors.get('gen')],
        df[factors.get('dys')]
    ]))
```

```python
df['Happiness Score'] = df['Happiness Score'].fillna(sum([
    df[factors.get('eco')],    
    df[factors.get('fam')],
    df[factors.get('health')],
    df[factors.get('free')],
    df[factors.get('trust')],
    df[factors.get('gen')],
    df[factors.get('dys')]
    ])
) 
```

```python
df.isna().sum()
```

<h2 style="color:#3d6fad"> Here we see that we decreased the number of missing values from Health & Econmy columns </h2>
